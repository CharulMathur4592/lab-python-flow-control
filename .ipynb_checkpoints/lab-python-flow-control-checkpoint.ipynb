In the last lab, you were starting an online store that sells various products. To ensure smooth operations, you developed a program that manages customer orders and inventory.

2. Instead of asking the user to input the name of three products that a customer wants to order, do the following:

    b. Add the product name to the "customer_orders" set.
    
    c. Ask the user if they want to add another product (yes/no).
    
    d. Continue the loop until the user does not want to add another product.

3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for  the products that were ordered (those in "customer_orders").

```python
# Step 1: Define the product list
products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
# Step 2: Create an empty dictionary
inventory = {}
```

```python
# Step 3: Ask the user for the quantity of each product
print("Enter the quantity for each product: ")
for item in products:
    quantity = int(input(f"Quantity of {item}: "))
    inventory[item] = quantity
```

```python
# Step 4: Create an empty set
customer_orders = set()
```

```python
#Step 5: Ask the user to enter the name of a product that a customer wants to order.

print("Available products: ", products)

while True:
    product = input("Enter the product that customer wants to order: ").strip().lower()

    while product not in products:
        print("Invalid Product. Choose from: ", products)
        product = input("Enter a valid product: ").strip().lower()

    customer_orders.add(product)

    another_product = input("Do you want to add another product? (yes/no): ").strip().lower()
    
    if another_product != "yes":
        break
```

```python
# Step 6: Print ordered products

print("Customer Orders: ", customer_orders)
```

```python
# Step 7: Calculate order statistics

total_products_ordered = len(customer_orders)
percentage_ordered = (total_products_ordered / len(products)) * 100

# Storing in tuple
order_status = (total_products_ordered, percentage_ordered)
```

```python
# Step 8: Print the statistics

print("Order statistics: ")
print("Total products ordered: ", order_status[0])
print("Percentage of products ordered: ", f"{order_status[1]:.2f}%")
```

```python
# Step 9: Update the inventory only for the ordered items

for item in customer_orders:
    if inventory[item] > 0:
        inventory[item] -= 1
```

```python
# Step 10:Updated inventory

print("Updated inventory: ")
for item, quantity in inventory.items():
    print(f"{item}: {quantity}")
```
